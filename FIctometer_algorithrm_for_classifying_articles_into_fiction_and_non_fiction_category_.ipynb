The Fictometer algorithm is based on the Part of Speech (POS) tags in a text. For a given input text, it first counts the number adverbs, adjectives, and pronounds in the text and uses it as an input to the Logistic Regression algorithm to do the classification.

We will be using the Brown corpus dataset for this work, and this corpus has both the text as well as the POS tags (added by human experts). However, the POS tags present in this corpus are finer, meaning adjectives can be further sub-divided into finer categories. But for our analysis, we only need the high level tags. And so the first step is to convert/group the finer tags into high level tags, which we do through the functions defined below.

```python
# now we define a function to count the number of POS tags in the text .

# This function counts the number of Upos tags and Pos tags .

def n_adj(text):
    adj=0
    for i in text :
        if i[0] == 'J':
          adj=adj+1
    return adj

# this function counts the number of nouns

def n_noun(text):
    noun = 0
    for i in text:
        if ((i[0] == 'N') and (i[1] != 'C')):
          noun=noun+1
    return noun

# this fuction counts the number of verbs
def n_verb(text):
    verb=0
    for i in text:
        if i[0] == 'V':
             verb=verb+1
    return verb

# This function counts the number  of pronouns

def n_pronoun(text):
    pronoun=0
    for i in text:
        if (i[0] == 'P') or (i[:3] in ['WP$','WPO','WPS']):
            pronoun=pronoun+1
    return pronoun

# This function counts the number of adverbs
def n_adv(text):
    adv=0
    for i  in text:
        if (i[0] == 'R') or (i[:3] in ['WRB']):
            adv=adv+1
    return adv

# this function outputs the universal high level tag using a finer tag as input

def func_utag(tag):
    if tag[0] == 'J' or tag == 'ADJ':
        utag='ADJ'
    elif ((tag[0] == 'N') and (tag[1] !='C')) or tag == 'NOUN':
         utag='NOUN'
    elif tag[0] == 'V' or tag == 'VERB':
         utag='VERB'
    elif (tag[0] == 'P') or (tag[:3] in ['WP$','WPO','WPS']) or tag == 'PRON':
         utag='ADV'
    else:
         utag='unknown'
    return utag

# this function outputs True or False Depending on whether the input tag
## is one  of the  five high level universal tag or not

def func_is5tag(tag):
    if tag in  ['ADJ','ADV','NOUN','PRON','VERB']:
         is5tag = True
    else:
        is5tag=False
    return is5tag
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
import pandas as pd

## define a fuction to calcutate pos tags
nltk.download('punkt')
nltk.download('wordnet')

nltk.download('averaged_perceptron_tagger')


def count_pos_tags(text):
    tokens = word_tokenize(text)
    pos_tags = nltk.pos_tag(tokens)
    pos_counts = {
        'Noun':0,
        'Verb':0,
        'Adjective':0,
        'Adverb':0
    }

    for word, pos in pos_tags:
        if pos.startswith('N'):
          pos_counts['Noun'] += 1
        elif pos.startswith('V'):
          pos_counts['Verb'] += 1
        elif pos.startswith('J'):
            pos_counts['Adjective'] += 1
        elif pos.startswith('R'):
          pos_counts['Adverb'] += 1

    return pos_counts







```

```python
import nltk
from nltk.corpus import wordnet as wn
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Downloading the wordnet dataset from nltk

nltk.download('wordnet')

## create a list of labeled exampes

'''labeled_data = [
    ('dog', 'animal'),
    ('cat', 'animal'),
    ('car', 'non-animal'),
    ('table', 'non-animal'),
    ('book','non-animal'),
    ('pencil','non-animal'),
    ('elephant','animal')
]'''
'''
labeled_data = {"fiction_words":[
    'fantasy', 'adventure', 'plot', 'character', 'novel', 'story', 'dialogue', 'imagination',
    'magical', 'mystery', 'narrative', 'setting', 'prose', 'genre', 'fictional'
],"non_fiction_words":[
    'information', 'fact', 'history', 'research', 'explanation', 'analysis', 'data', 'article',
    'essay', 'report', 'documentary', 'non-fiction', 'biography', 'reference', 'science'
]}'''

fiction_words = [
    'fantasy', 'adventure', 'plot', 'character', 'novel', 'story', 'dialogue', 'imagination',
    'magical', 'mystery', 'narrative', 'setting', 'prose', 'genre', 'fictional'
]

non_fiction_words = [
    'information', 'fact', 'history', 'research', 'explanation', 'analysis', 'data', 'article',
    'essay', 'report', 'documentary', 'non-fiction', 'biography', 'reference', 'science'
]

# Sample list of words to classify
words_to_classify = ['plot', 'history', 'character', 'research', 'adventure']

labeled_results = []

for word in words_to_classify:
    if word in fiction_words:
      label = 'fiction'
    elif word in non_fiction_words:
        label = 'non-fiction'
    else:
        label = 'unknown'
    labeled_results.append((word,label))

for word, label in labeled_results:
    print(f"Word: {word}| Label: {label}")

# define a fuction to extract hypernym features

def get_hypernym_features(word):
    features = {}
    synsets = wn.synsets(word)
    for synsets in synsets:
        hypernyms = synset.hypernyms()
        for hypernym in hypernyms:
            features[hypernym.name()] = 1
    return features

# Extract features and labels
X = [get_hypernym_features(word) for word, _ in labeled_results]
y = [label for _, label in labeled_results]

# conver the fearure to a matrix form

vectorizer = DictVectorizer()
X_matrix = vectorizer.fit_transform(X)

# split the data into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X_matrix,y,test_size=0.2, random_state=42)

# train a ml linear model

model = LogisticRegression()
model.fit(X_train, y_train)

# make prediction on the test data
y_pred = model.predict(X_test)

# calcutate accuracy
accuracy = accuracy_score(y_test,y_pred )
print("Accuracy:",accuracy)
```
